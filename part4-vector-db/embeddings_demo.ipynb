# Embeddings Demo

This notebook creates sentence embeddings for 10 sentences across Cricket, Cooking, and Cybersecurity, visualizes the cosine similarity matrix as a heatmap, and retrieves the top 2 matches for a new cricket query.

```python
sentences = [
    "The batsman scored a century after a patient opening stand.",
    "The bowler swung the new ball and dismissed two batters early.",
    "Fielders saved crucial boundaries in the final overs of the cricket match.",
    "The captain rotated the spinners well on a dry pitch.",
    "The recipe needs onions, tomatoes, and fresh spices for the curry.",
    "Baking bread requires steady heat and accurate measurements.",
    "The chef simmered the sauce until the vegetables turned tender.",
    "Strong passwords and multi-factor authentication protect user accounts.",
    "Phishing emails try to steal credentials by pretending to be trusted messages.",
    "Encryption helps secure sensitive files during storage and transmission."
]

query_sentence = "The bowler took three wickets in one over"
print(f"Loaded {len(sentences)} sentences across 3 topics.")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
similarity_matrix = cosine_similarity(embeddings)

plt.figure(figsize=(9, 7))
sns.heatmap(similarity_matrix, cmap='YlGnBu', annot=True, fmt='.2f')
plt.title('10x10 Cosine Similarity Heatmap')
plt.show()

query_embedding = model.encode([query_sentence])
query_scores = cosine_similarity(query_embedding, embeddings)[0]
top_indices = query_scores.argsort()[-2:][::-1]

print('Top 2 matches for the query:')
for rank, idx in enumerate(top_indices, start=1):
    print(f"{rank}. {sentences[idx]} -> {query_scores[idx]:.4f}")
```
